## Spectral integrals in Underworld2

Underworld2 currently supports area (volume) and surface integrals. But there is no inherent capacity to integrate a 2D function independently along separate axes. However, we can generate a Fourier representation of such an integral. We'll consider a 2D scalar field $F(x,y)$, over a domain with height $L$, width $W$. We want to determine the function representing the the integrated value over the y-axis: 


\begin{equation}f(x) = \int_0^L F(x,y) \mathop{dy}   \end{equation}

We can write this as a sum of periodic functions:


\begin{equation}f(x) = \frac{1}{W}  A_0  + \sum_K \frac{2}{W}  a_k \cos(kx) + \frac{2}{W}  b_k sin(kx) \end{equation} 

where:

\begin{equation} a_k = \int_0^L  \int_0^W \big( \mathop{F(x,y)} \cos(k x) \big)\mathop{dx}\mathop{dy} \end{equation} 


\begin{equation} b_k = \int_0^L  \int_0^W \big( \mathop{F(x,y)} \sin(k x) \big)\mathop{dx}\mathop{dy} \end{equation} 


\begin{equation} A_0 = \int_0^L  \int_0^W  \mathop{F(x,y)} \mathop{dx}\mathop{dy} \end{equation} 

The Underworld2 `function` and `utils` modules provides all the necessary apparatus us to compute these volume (area) integrals, and hence find the spectral coefficients  $A_0, a_k, b_k $.

A specific application of this procedure is the spectral domain calculation of gravity anomalies, also known as __upward continuation__. We'll cover some examples of upward continuation later in the tutorial. But first, let's consider a simple function in the unit square with an origin at (0,0):


\begin{equation} F(x,y) = \cos(\pi x) * \cos(\pi y)\end{equation}

\begin{equation}f(x) = \int F(x,y) \mathop{dy}  = \frac{2}{\pi} \sin(\pi x) \end{equation} 

\begin{equation} A_0 = \int_{-0.5}^{0.5} \int_{-0.5}^{0.5} F(x,y) \mathop{dx}\mathop{dy}  = \left(\frac{2}{\pi} \right)^2 \end{equation} 

## Set up Underworld model

```python
import underworld as uw
from underworld import function as fn
import glucifer
import numpy as np
from scipy.integrate import simps, trapz

%pylab inline
```

```python
# first create a mesh and variable
mesh = uw.mesh.FeMesh_Cartesian(elementType = 'Q1/dQ0', 
                                elementRes=(51,51),
                                minCoord    = (-0.5, -0.5), 
                                maxCoord    = (0.5, 0.5) )


```

```python
#Set up our example function
xFn = fn.input()[0] 
yFn = fn.input()[1]

yCosFn = (fn.math.cos(np.pi*yFn))
xCosFn = (fn.math.cos(np.pi*xFn))

combFn = xCosFn*yCosFn
```

```python
fig = glucifer.Figure(quality=3, figsize=(300,300))
fig.append( glucifer.objects.Surface(mesh,combFn, onMesh=True, colourBar=False ))
fig.append( glucifer.objects.Contours(mesh,combFn, colours = 'black',colourBar=False, interval=0.25) )

fig.show()
```

## Create the basis functions

Here we create the set of sin and cosine modes. The final UW functions are are vector functions of length (n - 1). 
We have to treat the zeroth mode ($A_0$) differently.

```python
# create set of wavenumbers / modes

res = mesh.elementRes[0]
width = mesh.maxCoord[0] - mesh.minCoord[0]

n =  11                #number of modes to use ( incl. the zeroth mode )

x = fn.coord()[0]
modes = []
ks_ = []
#start from 1, as the zeroth mode is performed using a regular volume integral)
for i in range(1,n):
    print(i)
    factor = float(i)*2.*np.pi/width
    modes.append(factor*x)
    ks_.append(factor)
sinfns = fn.math.sin(modes)
cosfns = fn.math.cos(modes)
```

## Compute the coefficients

```python
#Note that cosfns, sinfns are vector-valued. 
#The Integral is performed for each element, and a vector result wil be retruned

#a_k
cos_coeffs = uw.utils.Integral(combFn*cosfns,mesh)

#b_k
sin_coeffs = uw.utils.Integral(combFn*sinfns,mesh)

#A_0
average = uw.utils.Integral(combFn,mesh)


synthFn = (2./width)*fn.math.dot(sin_coeffs.evaluate(),sinfns) + \
          (2./width)*fn.math.dot(cos_coeffs.evaluate(),cosfns) + \
                     (1./width)*average.evaluate()[0]
 
#This is the analytic function (to be evaluated at top surface of the mesh)
trueFn = (2./np.pi)*xCosFn
```

```python
tWalls = mesh.specialSets["MaxJ_VertexSet"]
bWalls = mesh.specialSets["MinJ_VertexSet"]
lWalls = mesh.specialSets["MinI_VertexSet"]
```

```python
fig, ax = plt.subplots(figsize=(5, 3))

ax.plot(mesh.data[tWalls.data][:,0], synthFn.evaluate(tWalls ))
ax.plot(mesh.data[tWalls.data][:,0], trueFn.evaluate(tWalls ),ls='--' )
```

## Inspect the coefficients

```python
#evalate the intagrals - returns a vector result containing the a_k & b_k
sc = sin_coeffs.evaluate()
cc = cos_coeffs.evaluate()


oscill = np.ones(len(cc[:]))
oscill[0::2] *=-1
```

```python
fig, ax = plt.subplots(figsize=(10, 4))

ax.plot(np.arange(n-1) + 1, -1.*oscill*sc[:],label = 'amp. of sin modes', marker='o'  )
ax.plot(np.arange(n-1) + 1, -1.*oscill*cc[:], label = 'amp. of cosine modes' , marker='o' )

ax.legend()

ax.hlines(0,0, n, linestyles='--', lw=1. )
#ax.vlines(nk - 1,-0.2*cc[0], cc[0], linestyles='--', lw=1. )

ax.set_xlabel('N')
ax.set_ylabel('amplitude')
```

```python
meshxs = mesh.data[tWalls.data][:,0]
width = abs(mesh.maxCoord[0] - mesh.minCoord[0])
sum_ = np.zeros(len(meshxs))
for i in range(len(cc) - 1):
    #print(i)
    plt.plot(meshxs, cc[i]*np.cos(ks_[i]*meshxs), c='k', ls = '--' )
    sum_ +=cc[i]*np.cos(ks_[i]*meshxs)
    
plt.plot(meshxs, sum_, c= 'b')    
```

## Comparison with Numpy Version (& mode leakage)

Although the analytic function ($f(x)$) contained a single wavenumber, the Underworld integral (spectral integral) contained finite amplitudes at a series of wavenumbers. This is a case of spectral leakage, a well-known phenemonon caused by sampling. The degree of spectral leakage will depend on the domain size chosen. Try setting the domain width to 2, and the leakage should vanish.


We'll now undertake the same proceedure, using Numpy tools. First we'll create the row-wise FFT, then we'll sum vertically to generarate $f(x)$.Because Numpy uses the FFT, we can use all wavenumbers at very little computational cost (but is restricted to running on a single processor). The spectral leakage is very similar. 

```python
#reshape our function into 2D array for Numby to work with
arrayData = combFn.evaluate(mesh).reshape(mesh.elementRes[1] + 1, mesh.elementRes[0] + 1)

fig, ax = plt.subplots()
ax.contourf(arrayData)
ax.set_aspect('equal')
```

```python
#compute the numpy FFT 

arrayFFT = np.fft.rfft(arrayData) #create the row-wise FFT 

depths = mesh.data[lWalls.data][:,1]
fftSum = simps(arrayFFT, x = depths, axis=0) #integrate in y using Simsons' Rule

#do the inverse transform (Fourier synthesis 
arraySum = np.fft.irfft(fftSum)
```

```python
arrayXs = np.linspace(mesh.minCoord[0], mesh.maxCoord[0],  arraySum.shape[0])


fig, ax = plt.subplots(figsize=(5, 3))

ax.plot(arrayXs , arraySum)
ax.plot(mesh.data[tWalls.data][:,0], trueFn.evaluate(tWalls ),ls='--' )
```

```python

N = arrayData.shape[0]

fig, ax = plt.subplots(figsize=(10, 4))

ax.plot(-1.*np.real_if_close(fftSum)[1:20]/N, marker = 'o', label = 'modes using numpy'  )
ax.plot(-1.*oscill*np.array(cc[:]), marker = 'o',label = 'modes using Underworld integrals' )

ax.hlines(0,0, n, linestyles='--', lw=1. )


ax.legend()
ax.set_xlabel('N')
ax.set_ylabel('amplitude')
```
